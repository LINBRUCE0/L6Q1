```python
import pandas as pd
watchshow = pd.read_csv('watch.csv')
watchshow
```

```python
x = watchshow.drop(columns=['watch'])
x
```

```python
y = watchshow['watch']
y
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x,y)
prediction = model.predict([[30,1]])
prediction
```

```python
import joblib
joblib.dump(model,'watch-recommender.joblib')
```
